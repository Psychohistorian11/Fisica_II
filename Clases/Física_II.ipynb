##  Movimiento  periodico  de un  sistema  masa  resorte

![](https://colcafe.files.wordpress.com/2013/05/resorte_oscilante.png)

El  primer  sistema que  estudiaremos es un  sistema ccompuesto  por  una massa  y  un  resorte.  Por simplicidad  lo  tomaremos en  posición  hoirizontal.
Como  tenemos  el  diagrama de cuerpo  libre,  nuestro  siguiente paso  es  escribir la segunda ley  de Newton

---

$$m\frac{d^{2}x(t)}{dt^2}=-kx(t) \tag{1}$$


---

Todo  nuestro  problema  se  reduce  a resolver  la  ecuación  diferencial.  para esto  podemos entrar  en  panico  o  aplicar  lo  que  sabemos al  momento

Para  la cual  la solución es

$$x(t)=A\cos{(\omega t +\phi)}$$

```python
import numpy as np
import matplotlib.pyplot as plt

k = 200   #N/m
m = 0.2   #Kg
phi = 0.0 #Condiciones iniciales
A = 0.5   #Amplitud en m
w = np.sqrt(k/m)
T = 2*np.pi * np.sqrt(m/k)


t = np.arange(0.,3*T,3 * T/200)


x = A*np.cos(w*t+phi)
v = -A*w*np.sin(w*t+phi)
a = -A* w**2*np.cos(w*t+phi)


plt.plot(t,x)
plt.xlabel("Tiempo(s)")
plt.ylabel("x(t)")
plt.show()
plt.plot(t,v,color="red")
plt.show()
plt.plot(t,a,color="green")
```

## Solución  analitica



```python
from sympy  import *

x = symbols('x', cls=Function)
A,t,k,m =symbols('A t k m')
dx = diff(x(t),t,2)
dsolve(dx+k*x(t)/m,x(t),ics={x(0): A, diff(x(t), t).subs(t,0): 0})
```
