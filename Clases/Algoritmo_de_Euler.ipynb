## Metodo  de Euler

En  una dimensión  en la forma más general  podemos escribir la  segunda ley  de Newton como

$$m\frac{d^2x(t) }{dt^2}=F_{x}$$

Podemso  aplicar las definiciones   generales, esto  es

$$v(x)=\frac{dx(t)}{dt}$$

$$a(t)=\frac{dv(t)}{dt}$$


Emplearemos la aproximación de Euler pra  tratar  con estas cantidades


$$\frac{df(x)}{dx}≈ \frac{f(x+\Delta x)-f(x)}{\Delta x}
$$

aplicando  esto  a la  aceleración  tenemos que

$$a(t)≈ \frac{v(t+\Delta t )-v(t)}{\Delta t}$$

por tanto de laa gundd ey  de Neton nos queda que

$$m\frac{v(t+\Delta t )-v(t)}{\Delta t} =F_{x}$$

$$v(t+\Delta t) \approx v(t)+\frac{\Delta t}{m}F_{x}$$
ulitizando la definición  de velocidad  tenemos que


$$v(t) \approx \frac{x(t+\Delta t)-x(t)}{\Delta t}$$
finalmente

$$x(t+\Delta t)\approx x(t)+v(t)\Delta t$$

#  Algoritmo  de  Euler

Este  me sirve para  resolver las ecuaciones  de movimiento  de un  sistema físico.  En  el  caso  de 1D,  tenemos  

$$v(t+\Delta t)\approx v(t)+\frac{\Delta t}{m}F_{x}(t)$$

$$x(t+\Delta t)\approx x(t)+v(t)\Delta t$$

en  terminos de indices  esto  lo  podemos escribir

$$v_{i+1}\approx v_{i}+\frac{\Delta t}{m}F_{i}$$

$$x_{i+1}\approx x_{i}+v_{i}\Delta t$$

Para  una solución  particular  debemos definir  las condiciones iniciales  del  problema,  es decir debemos  tener  $x_{0}$  y  $v_{0}$

```python
def eulaer1d(x0,v0,m,k,dt=0.000001,nst=2000000):
  """
  x0 :  Is the initial position
  v0 :  Is the initial velocity
  m  :  Is the mass of the object
  k  :  Spring constant of elasticity
  dt :  t differential
  nst:  Number of steps
  """
  x=[]
  v=[]
  t=[]
  x.append(x0)
  v.append(v0)
  t.append(0.0)
  for i  in  range(nst):
    v.append(v[i]-dt*k*x[i]/m)
    x.append(x[i]+v[i]*dt)
    t.append(t[i]+dt)
  return t,x,v


#system initial conditions
#------------------------------------------------
m=0.1
k=200
x0 =0.5
v0 =0.0
#-------------------------------------------------


t,x,v = eulaer1d(x0,v0,m,k)
```

```python
import matplotlib.pyplot as plt

#Graficar
plt.plot(t,x,color="green")
plt.xlabel("Tiempo(s)")
plt.ylabel("x(t)")
plt.show()
plt.plot(t,v,color="yellow")
plt.xlabel("Tiempo(s)")
plt.ylabel("v(t)")
```
